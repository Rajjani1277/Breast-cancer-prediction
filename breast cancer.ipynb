# Feature Selection

```python
def getfeatures(data, t):
    
    feature = []
    value = []
    
    for i,index in enumerate(data.index):
        if abs(data[index]) > t:
            feature.append(index)
            value.append(data[index])
            
    df = pd.DataFrame(data = value, index = feature, columns=['corr value'])
    return df
```
